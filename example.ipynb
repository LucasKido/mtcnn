```python
import numpy as np
import cv2
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
```

```python
img = cv2.cvtColor(cv2.imread('henry_cavill.jpg'), cv2.COLOR_BGR2RGB)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(img)
ax.axis('off')
plt.savefig('henry_cavill.png', pad_inches=0.1)
```

```python
detector = MTCNN()
result = detector.detect_faces(img)
```

```python
result
```

```python
x1, y1, width, height = result[0]['box']
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(img)
ax.add_patch(Rectangle((x1, y1), width, height, fill=False, ec='r', lw=2))
ax.axis('off')
plt.savefig('henry_bbox.png', pad_inches=0.1)
plt.show()
```

```python
left_eye, right_eye = result[0]['keypoints']['left_eye'], result[0]['keypoints']['right_eye']
mouth_left, mouth_right = result[0]['keypoints']['mouth_left'], result[0]['keypoints']['mouth_right']
nose = result[0]['keypoints']['nose']
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(img)
ax.add_patch(Rectangle((x1, y1), width, height, fill=False, ec='r', lw=2))
ax.add_patch(Circle(left_eye, fill=True, color='red', radius=15))
ax.add_patch(Circle(right_eye, fill=True, color='red', radius=15))
ax.add_patch(Circle(mouth_left, fill=True, color='red', radius=15))
ax.add_patch(Circle(mouth_right, fill=True, color='red', radius=15))
ax.add_patch(Circle(nose, fill=True, color='red', radius=15))
ax.axis('off')
plt.savefig('henry_landmark.png', pad_inches=0.1)
plt.show()
```
